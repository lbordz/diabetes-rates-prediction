```python
base = 'LI:YEAR:'                                           
```

```python
new_col_names = ['', '', '']
for year in range (2004, 2014):
    new = [base.replace('YEAR', str(year))] * 7
    new_col_names = new_col_names + new
```

```python
#checking aligment of my new column names to original ones
pd.DataFrame(new_col_names, np.array(li_years.columns)).tail(10)
```

```python
x = np.array(new_col_names + li_data.columns)
for i in range(10, len(x)):
    x[i] = x[i][:-2]
x
```
