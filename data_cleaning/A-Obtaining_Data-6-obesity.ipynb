```python
import pandas as pd
import numpy as np
```

```python
#1  Importing entire dataset
```

```python
ob_data = pd.read_excel('../data/raw_data/OB_PREV_ALL_STATES.xlsx', skiprows=1)  #
ob_years = pd.read_excel('../data/raw_data/OB_PREV_ALL_STATES.xlsx', nrows=1)
```

```python
ob_data.head()
```

```python
ob_years
```

```python
base = 'OB:YEAR:'                                           
```

```python
new_col_names = ['', '', '']
for year in range (2004, 2014):
    new = [base.replace('YEAR', str(year))] * 7
    new_col_names = new_col_names + new
```

```python
#checking aligment of my new column names to original ones
pd.DataFrame(new_col_names, np.array(ob_years.columns)).tail(10)
```

```python
ob_data.columns
```

```python
new_col_names_long = np.array(new_col_names + ob_data.columns)
```

```python
for i in range(10, len(new_col_names_long)):
    new_col_names_long[i] = new_col_names_long[i][:-2]
```

```python
ob_data.columns = new_col_names_long
```

```python
ob_data.head()
```

```python
ob_data.columns
```

## Picking desired columns  #Second pass!
Second pass! Obesity Rate 2010 (first pass) + Change in obesity 2009-2010

```python
ob2 = ob_data[['State', 'FIPS Codes', 'County', 'OB:2010:percent', 'OB:2009:percent']]
```

```python
ob2.info()
```

```python
ob2[ob2[['OB:2010:percent', "OB:2009:percent"]].applymap(np.isreal).sum(axis = 1) < 2]
```

```python
ob2['OB:2010:percent'] = pd.to_numeric(ob2['OB:2010:percent'], errors = 'coerce')
ob2['OB:2009:percent'] = pd.to_numeric(ob2['OB:2009:percent'], errors = 'coerce')
```

```python
ob2['OB:2009-2010:Change_rate'] = ( ob2['OB:2010:percent'] - ob2['OB:2009:percent'] ) / ob2['OB:2009:percent'] 

```

```python
ob2.head()
```

```python
ob2.info()
```

```python
ob2[ob2['OB:2010:percent'].isnull()].head()
```

```python
ob2 = ob2[ob2.isnull().sum(axis = 1) == 0]
```

```python
ob2.info()
```

```python
ob2 = ob2.drop('OB:2009:percent', axis = 1)
```

```python
# Exporting to CSV  (second-pass)

ob2.to_csv('../data/Obesity_data_2.csv')
```
