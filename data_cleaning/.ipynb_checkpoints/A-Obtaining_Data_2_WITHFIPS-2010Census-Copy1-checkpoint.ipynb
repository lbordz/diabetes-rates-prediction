```python
fips['StateCounty'] = fips['state'] + fips['census_CTYNAME']
df['StateCounty'] = df["STNAME"] + df["CTYNAME"]
df.shape
```

```python
newdf = df.merge(fips[['StateCounty', "fips", "state", "county", 'census_CTYNAME']], on = "StateCounty", how = 'left')
newdf.head()
```

```python
## Only taking Year = 1
```

```python
starter_demos = newdf[['STNAME', 'CTYNAME', "fips", "YEAR", "AGEGRP", "TOT_POP", "TOT_MALE", "TOT_FEMALE"]]
```

```python
starter_demos["pct_male"] =  starter_demos["TOT_MALE"] / starter_demos["TOT_POP"]
```

```python
first_pass = starter_demos[starter_demos["YEAR"] == 1]   #Year 1 is 2010, data from the original sensus
first_pass.head()
```

```python
Age_Mapping = {
    1: "2010:AgeGroup01:0-4",
    2: "2010:AgeGroup02:5-9",
    3: "2010:AgeGroup03:10-14",
    4: "2010:AgeGroup04:15-19",
    5: "2010:AgeGroup05:20-24",
    6: "2010:AgeGroup06:25-29",
    7: "2010:AgeGroup07:30-34",
    8: "2010:AgeGroup08:35-39",
    9: "2010:AgeGroup09:40-44",
    10: "2010:AgeGroup10:45-49",
    11: "2010:AgeGroup11:50-54",
    12: "2010:AgeGroup12:55-59",
    13: "2010:AgeGroup13:60-64",
    14: "2010:AgeGroup14:65-69",
    15: "2010:AgeGroup15:70-74",
    16: "2010:AgeGroup16:75-79",
    17: "2010:AgeGroup17:80-84",
    18: "2010:AgeGroup18:85+",
}
```

```python
first_pass.head()
```

```python
first_pass_age_totals = first_pass[first_pass["AGEGRP"] == 0][["STNAME", "CTYNAME", "fips", "AGEGRP", "TOT_POP", "pct_male" ]].sort_values(by=['fips'])
first_pass_age_totals.head()
```

```python
for i in range(1, 19):
    agegrp = first_pass[first_pass["AGEGRP"] == i][["fips", "AGEGRP", "TOT_POP" ]].sort_values(by=['fips'])
    print(agegrp.head())
    first_pass_final[Age_Mapping[i]] = np.array(agegrp["TOT_POP"]) / np.array(first_pass_age_totals["TOT_POP"])
    print(first_pass_final[Age_Mapping[i]].head())
```
