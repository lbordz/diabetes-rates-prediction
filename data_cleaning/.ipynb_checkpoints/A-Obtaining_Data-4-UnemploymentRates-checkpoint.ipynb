```python
import pandas as pd
import numpy as np
```

```python
from os import listdir
from os.path import isfile, join
```

```python
filenames = [f for f in listdir('../data/raw_data/Unemployment/')]
```

```python
filenames.remove('.DS_Store')
```

```python
filenames[0]
```

```python
alldfs = []
for filename in filenames:
    df = pd.read_excel('../data/raw_data/Unemployment/' + filename, skiprows = 1) 
    #rename columns and reorg them
    beg = ["FIPS"]
    beg2 = [df.columns[1]]
    middle = list(df.columns[2:12].astype(str) + [":UnemploymentRate"])
    end = [df.columns[12]]
    endend = [df.columns[-1] + " (2017)"]
    df.columns = beg + beg2 + middle + end + endend
    #drop extra columns that mean nothing
    df.drop("None:UnemploymentRate", axis = 1, inplace = True)
    #drop first row, which is always the total for the STATE
    df.drop(index = 0, inplace = True)
    alldfs.append(df)    
```

```python
master_df = pd.DataFrame(columns = alldfs[0].columns)
for df in alldfs:
    master_df = master_df.append(df)
```

```python
master_df.tail(10)
```

```python
master_df = master_df[master_df['Name'].isnull() == False]
master_df.tail()
```
